## Базовые статистические характеристики данных

Сейчас будет немного определений из ТВиМС, которые вы вероятно знаете. Но это нужно, чтобы сопоставить один к одному.

$\bullet$ Наш набор данных называется **случайной выборкой**.

$\bullet$ Один столбец называется **случайной величиной**.

У случайной величины есть основные статистические характеристики, которые применяются для предварительного анализа данных. Возьмем в рассмотрение случайную величину
$$X = \{Fare\} = \{x_1,\ldots,x_N\},$$
то есть наш столбец из датафрейма:

```python
X = data['Fare']
X
```

Для этой случайной величины мы можем посмотреть описательные статистики. Сначала мы рассмотрим конкретные команды для основных описательных статистик, а потом одну команду, которая выводит все общие описательные статистики. 

**Математическое ожидание**.
В условиях случайной выборки математическое ожидание равносильно арифметическому среднему всех значений случайной величины.
$$\mu = \mathbf EX = \dfrac1N\sum\limits_{i=1}^Nx_i$$
Вычислить среднее можно с помощью следующей команды:

```python
X.mean()
```

**Мода.**
Значение случайной величины, которое встречается чаще всего
$$m = \underset{i=1,\ldots,N}{\operatorname{argmax}}P(X = x_i)$$
Вычислить моду можно с помощью следующей команды:

```python
X.mode()
```

**Медиана.** Среднее по порядку значение 
$$M=F_X^{-1}(0.5),$$
где $F_X$ -- функция распределения СВ $X$.

Вычислить медиану можно с помощью следующей команды:

```python
X.median()
```

**Стандартное (среднеквадратичное) отклонение**. Отклонение случайной величины от среднего значения
$$\sigma = \sqrt{DX} = \sqrt{\dfrac{\sum\limits_{i=1}^N (x_i - \mu)^2}{N}}$$
Вычислить стандартное отклонение можно с помощью следующей команды:

```python
X.std()
```

Есть также и остальные описательные статистики. С помощью команды describe мы можем вывести сразу большинство из них (количество, среднее, отклонение, минимальное значение, 1/4-квантиль, 1/2-квантиль, 3/4-квантиль, максимальное значение)

```python
X.describe()
```

Мы также можем обобщить это на всю нашу случайную выборку, используя следующую команду

```python
data.describe()
```

## Проверка гипотез

Если встает вопрос о каких-то более глубоких статистических исследованиях, как например, проверка гипотез, то для этого хорошо подходит библиотека SciPy (https://scipy.org), в которой находится много классов. Нас в данном случае интересует класс stats.

```python
from scipy import stats
```

Итак, возьмем самый простейший случай проверки гипотез, а именно через $t$-критерий Стъюдента. Освежим в памяти, что это метод статистической проверки гипотез о равенстве средних двух выборок, основанных на распределении Стъюдента.

Пусть у нас имеются две выборки из нормального распределения с одинаковой дисперсией $$X \sim \mathcal N (1, 2),\ Y \sim \mathcal N(0.5, 2)$$
(поскольку $t$-критерий корректно работает только с нормально распределенными выборками).

```python
X = np.random.normal(loc=0, scale=2, size=150)
Y = np.random.normal(loc=0.5, scale=2, size=150)
```

Мы уже умеем строить гистограммы, поэтому посмотрим как выглядят гистограммы, чтобы убедиться в том, что у нас выборки действительно из нормального распределения:

```python
sns.histplot(X)
sns.histplot(Y)
plt.legend(['X', 'Y'])
```

Как можно видеть, гистограммы действительно напоминают нормальное распределение. 

Теперь выдвинем гипотезу
$$H_0 : \{\mu_1 = \mu_2\},$$
где $$\mu_1 = \dfrac{1}{N}\sum\limits_{i=1}^N x_i,\ \mu_2 = \dfrac{1}{N}\sum\limits_{i=1}^N y_i,$$
а вместе с ней альтернативную гипотезу
$$H_1 : \{\mu_1 \ne \mu_2\}.$$

С помощью $t$-критерия Стъдента проверим, можно ли принять гипотезу $H_0$. Суть в том, что мы задаем критическое значение $\varepsilon$ для статистического критерия (обычно $\varepsilon = 0.05$) и проверяем превосходит ли Р-значение критерия критическое значение. Если Р-значение критерия выше $\varepsilon$, то мы не можем отклонить гипотезу $H_0$. Иначе, в случае, когда Р-значение ниже $\varepsilon$, мы можем отклонить гипотезу $H_0$.

$t$-критерий Стъдента можно задать следующим образом:

```python
t_statisitc, p_value = stats.ttest_ind(X, Y)
```

```python
epsilon = 0.05
p_value > epsilon
```

То есть Р-значение статистического критерия ниже порогового значения. Следовательно, мы можем отклонить гипотезу $H_0$ и принять гипотезу $H_1$ о несовпадении средних двух выборок. 

## Целостность данных

Очень часто в анализе данных приходится сталкиваться с проблемой целостности данных. Иногда некоторые случайные величины могут иметь пропущенные значения.Одна из целей предварительного анализа данных -- выявлять такие моменты и грамотно от них избавляться, дабы это не сказалось на последующем моделировании.

Вернемся к нашему датафрейму Titanic. Посмотреть общую информацию о всех случайных величинах (столбцах) датафрейма можно следующим образом:

```python
data.info()
```

Можно заметить, что в столбцах Age, Cabin, Embarked у нас не хватает значений. Однако при большом количестве столбцов заметить подобное может быть не так лекго, поэтому с помощью одной команды мы можем явно посмотреть, сколько данных в каких столбцах пропущено

```python
data.isna().sum()
```

Итак, мы явно видим, где у нас пропущены значения. Проще всего избавиться от пропущенных значений можно с помощью следующей команды

```python
data.dropna()
```

То есть мы получили новый датафрейм, в котором мы отбросили все объекты с пустыми значениями. Однако есть одна маленькая проблема -- мы потеряли львиную долю всех наших данных:

```python
data.dropna().shape
```

Теперь у нас 183 объекта, а было 891. Поэтому такой способ слишком радикальный и не очень полезный, хотя рабочий.

Поэтому мы поступим по-другому. Не будем удалять пропущенные значения, а будем их заменять на некоторые другие. Например, в графе Age, если возраст не указан, то мы можем взять среднее и заполнить им пропущенные значения. Пропущенные значения в столбце Cabin мы можем заменить на какую-нибудь строку, обозначающую то, что номер кабины не указан, например, "NO". Аналогично можно поступить и в столбце Embarked. Все это можно сделать следующим образом

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Cabin'] = data['Cabin'].fillna('NO')
data['Embarked'] = data['Embarked'].fillna('NO')
```

Теперь посмотрим, есть ли пропущенные значения в датафрейме

```python
data.isna().sum()
```

Как можно видеть, пропущенных значений не осталось. Наша цель выполнена.
