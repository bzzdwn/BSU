## Задание 1: Использование основных типов графиков.
Используйте набор данных о продажах автомобилей (https://www.kaggle.com/nehalbirla/vehicle-dataset-from-cardekho) и создайте различные типы графиков, чтобы отобразить различные аспекты данных.

Включите
- столбчатые диаграммы,
- гистограммы,
- линейные графики,
- диаграммы размаха
- ящик с усами
- радиальные диаграммы

```python
data = pd.read_csv('data/car data.csv')
data.head()
```

Построим столбчатую диаграмму для признака $Car\_Name$

```python
plt.figure(figsize=(10,16))
sns.countplot(y=data['Car_Name'])
plt.title('Cars count')
plt.show()
```

Из данного графика можно сделать вывод, что в наших данных присутствует больше всего автомобилей марки city.

Построим гистограмму распределения признака Selling_Price

```python
plt.figure(figsize=(10,6))
sns.histplot(x=data['Selling_Price'])
plt.title('Selling price histplot')
plt.show()
```

Из гистограммы можно сделать вывод, что распределение случайной величины напоминает экспоненциальное распределение.

Построим линейный график для временного ряда, который можно получить, взяв признаки Year и Selling_Price

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=data['Year'], y=data['Selling_Price'])
plt.title('Mean selling price Time Series')
plt.show()
```

Из данного графика можно понять, что временной ряд имеет тренд.

Построим boxplot для признака Selling_Price

```python
plt.figure(figsize=(10,6))
sns.boxplot(data['Selling_Price'])
plt.title('Selling price boxplot')
plt.show()
```

Из boxplot можно увидеть, что в данных присутствуют аномально большие значения. Это же мы наблюдали на гистограмме. Однако для экспоненциально распределенных случайных величин свойственно такое поведение.

Построим радиальную диаграмму для Car_Name и Selling_Price

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16,10))
plt.pie(data['Car_Name'].value_counts()[:20], labels=data['Car_Name'].value_counts()[:20].index, autopct='%1.1f%%')
plt.title('Cars Pie Chart')
plt.show()
```

Построим диаграмму рассения между признаками Present_Price и Selling_Price

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data[['Present_Price', 'Selling_Price']], x='Selling_Price', y='Present_Price')
plt.title('Prices scatterplot')
plt.show()
```

## Задание 2: Создание тепловой карты корреляции.
Используйте набор данных о погоде (https://www.kaggle.com/selfishgene/historical-hourly-weather-data) и с помощью библиотеки seaborn создайте тепловую карту, чтобы отобразить корреляцию между различными параметрами погоды.


```python
temperature = pd.read_csv('data/temperature.csv', parse_dates=['datetime'], index_col=0)
temperature.head()
```

```python
wind_speed = pd.read_csv('data/wind_speed.csv', parse_dates=['datetime'], index_col=0)
wind_speed.head()
```

```python
wind_direction = pd.read_csv('data/wind_direction.csv', parse_dates=['datetime'], index_col=0)
wind_direction.head()
```

```python
pressure = pd.read_csv('data/pressure.csv', parse_dates=['datetime'], index_col=0)
pressure.head()
```

```python
humidity = pd.read_csv('data/humidity.csv', parse_dates=['datetime'], index_col=0)
humidity.head()
```

Будем строить корреляционную матрицу для температуры в Портланде.

```python
portland_temperature = pd.DataFrame([temperature['Portland'], wind_speed['Portland'], wind_direction['Portland'], pressure['Portland'], humidity['Portland']], index=['temperature', 'wind_speed', 'wind_direction', 'pressure', 'humidity']).T
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(portland_temperature.corr(), annot=True)
plt.title('Portland Temperature heatmap')
```

Из корреляционной матрицы видно, что temperature имеет сильную отрицательную корреляцию с humidity.

## Задание 3: Использование plotly для создания интерактивных графиков.
Скачайте набор данных о выборах (например, https://www.kaggle.com/datasets/tunguz/us-elections-dataset) и используйте Plotly для создания интерактивных графиков, таких как интерактивные карты или диаграммы с подсказками.


```python
data = pd.read_csv('data/1976-2020-president.csv')
data.head(12)
```

```python
import plotly.express as px
```

Построим график количества голосов по штатам, используя цвет для визуализации.

```python
fig_map = px.choropleth(
    data,
    locations='state_po',
    locationmode='USA-states',
    color='candidatevotes',
    hover_name='candidate',
    hover_data=['party_simplified'],
    scope='usa',
    title='Голоса по штатам'
)

fig_map.update_layout(
    geo=dict(bgcolor= 'rgba(0,0,0,0)'),
    template='plotly'
)

fig_map.show()
```

Построим диаграммы количества голосов, полученных каждой партией в каждом штате.

```python
fig_bar = px.bar(
    data,
    x='state',
    y='candidatevotes',
    color='party_simplified',
    hover_data=['candidate'],
    title='Голоса по партиям'
)

fig_bar.update_layout(
    template='plotly'
)
```

# Задание 4: Создание визуализаций временных рядов.
Возьмите набор данных о COVID-19 от Джона Хопкинса (https://github.com/CSSEGISandData/COVID-19) и создайте визуализации временных рядов для показа изменения числа случаев заражения и смертей по времени.


```python
confirmed_data = pd.read_csv('data/time_series_covid19_confirmed_global.csv')
confirmed_data.head()
```

```python
confirmed_data.iloc[23]
```

```python
deaths_data = pd.read_csv('data/time_series_covid19_deaths_global.csv')
deaths_data.iloc[23]
```

```python
recovered_data = pd.read_csv('data/time_series_covid19_recovered_global.csv')
recovered_data.iloc[23]
```

```python
confirmed_ts = confirmed_data.iloc[23][4:].rename('confirmed')
deaths_ts = deaths_data.iloc[23][4:].rename('deaths')
recovered_ts = recovered_data.iloc[23][4:].rename('recovered')
```

```python
covid_data = pd.concat([confirmed_ts, deaths_ts, recovered_ts], axis=1)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(covid_data)
plt.legend()
plt.grid(False)
plt.show()
```

# Задание 5: Использование географической визуализации.

С использованием набора данных о террористических атаках (https://www.kaggle.com/START-UMD/gtd) создайте географическую визуализацию для показа, где происходят террористические атаки по миру, используя библиотеку Folium или Geopandas.

```python
import folium
```

```python
data = pd.read_csv('data/globalterrorismdb_0718dist.csv', delimiter=',', encoding='latin-1')
```

```python
data.head()
```

```python
m = folium.Map(location=[20, 0], zoom_start=2)

for idx, row in data[:100].iterrows():
    if pd.notnull(row['latitude']) and pd.notnull(row['longitude']):
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=f"{row['city']}, {row['country']}",
        ).add_to(m)

m.save('terrorism_map.html')
```
