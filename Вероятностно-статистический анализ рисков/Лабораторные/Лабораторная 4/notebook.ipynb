```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv', index_col=0).map(lambda x: float(x.replace(' ', '').replace(',', '.')))
data.head()
```

```python
x = data['Лукойл']
t = np.arange(-len(x.index)/2, len(x.index)/2)
```

```python
plt.figure(figsize=(20,10))
plt.plot(x)

# Add title and axis labels
plt.title('Time Series Plot')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
a = np.mean(np.dot(x.values, t)) / np.var(t)
b = np.mean(x.values) - a * np.mean(t)
```

```python
def line_trend(t):
    return a*t + b
```
