```python
import math
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
def f(x):
    return 2*np.sin(3*x)-(x**2-4*x+3)

def derivative_f(x):
    return 6*np.cos(3*x)-(2*x-4)

def second_derivative_f(x):
    return -18*np.sin(3*x) - 2
```

```python
x = np.linspace(-10, 10, 100)

fig, ax = plt.subplots()
ax.plot(x, 2*np.sin(3*x), label='y=2sin(3x)')
ax.plot(x, x**2-4*x+3, label='y=x**2-4*x+3')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.legend()
plt.grid()
plt.show()
```
