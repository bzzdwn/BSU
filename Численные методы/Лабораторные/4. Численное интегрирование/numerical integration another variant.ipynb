Необходимо вычислить значения $A_i$, $x_i$, $i=1,2,3$. Для этого построим систему из 6 уравнений
$$\begin{cases}

	x^1 : \int\limits_{2}^3 \dfrac{x}{\sqrt{(x-2)(3-x)}} dx = A_0x_0 + A_1x_1+ A_2x_2,\\
	x^2 : \int\limits_{2}^3 \dfrac{x^2}{\sqrt{(x-2)(3-x)}} dx = A_0x_0^2 + A_1x_1^2 + A_2x_2^2,\\

	x^5 : \int\limits_{2}^3 \dfrac{x^5}{\sqrt{(x-2)(3-x)}} dx = A_0x_0^5 + A_1x_1^5 + A_2 x_2^5.
	\end{cases}$$ 
Тогда
$$\begin{cases}
\pi = A_0 + A_1 + A_2,\\
\dfrac{5\pi}2 = A_0x_0 + A_1x_1+ A_2x_2,\\
\dfrac{51\pi}8= A_0x_0^2 + A_1x_1^2 + A_2x_2^2,\\

\dfrac{5603\pi}{128} = A_0x_0^4 + A_1x_1^4 + A_2 x_2^4,\\
\dfrac{30075\pi}{256} = A_0x_0^5 + A_1x_1^5 + A_2 x_2^5.
	\end{cases}$$ 
Эта система имеет решение, которое можно найти с помощью Wolfram Mathematica

$$A_0 = A_1 = A_2 = \dfrac \pi 3,\ x_0 = \dfrac14(10 - \sqrt3),\ x_1 = \dfrac52, x_2 =\dfrac14(10 + \sqrt3).$$

# Задача 3

3. Используя правило Рунге, провести сравнительный анализ квадратурных формул правых прямоугольников и трапеций на примере вычисления интеграла $$I = \int\limits_1^3 \dfrac{\lg x}{x^2+1} dx.$$

```python
a, b = 1, 3
def f(x):
    return np.log10(x) / (x**2 + 1)
```

```python
def runge_rule(m, compound_quadrature_formula, f, a, b, epsilon, N_divisions=False):
    h_1 = b-a
    h_2 = h_1 / 2
    iterations = 1
    I_h1 = compound_quadrature_formula(f, a, b, h_1)
    I_h2 = compound_quadrature_formula(f, a, b, h_2)
    R = (I_h2 - I_h1) / (1 - (h_2/h_1)**m)
    I = I_h1 + R
    while abs(R) >= epsilon:
        h_1 = h_2
        h_2 = h_1 / 2
        I_h1 = compound_quadrature_formula(f, a, b, h_1)
        I_h2 = compound_quadrature_formula(f, a, b, h_2)
        R = (I_h2 - I_h1) / (1 - (h_2/h_1)**m)
        I = I_h1 + R
        iterations += 1
    if N_divisions:
        return I, iterations
    return I
```

## Сравнение составных квадратурных формул

Теперь определим составные квадратурные формулы, которые мы будем использовать для вычисления.


$\bullet$ *Формула $$I_\text{пс}(f) = h\sum_{k=0}^{N-1}f(x_{k+1}) = h\sum_{k=0}^{N-1}f(a+(k+1)h)$$ называется **составной квадратурной формулой правых прямоугольников**.*

Она имеет алгебраическую степень точности равную 1.

Реализуем программно метод, вычисляющий эту сумму при заданных параметрах $f(x), a, b, h$.

```python
def right_rectangles(f, a, b, h):
    I = 0
    N = int((b-a)/h)
    for k in range(N):
        I += f(a + (k+1)*h)
    I *= h
    return I
```

$\bullet$ *Формула $$I_\text{тс}(f) = \dfrac{h}{2}\sum_{k=0}^{N-1}(f(x_k) + f(x_{k+1})) = h\left[\dfrac{f(a)+f(b)}{2} + \sum_{k=0}^{N-1}f(x_k)\right]$$ называется **составной квадратурной формулой трапеций**.*

Она имеет алгебраическую степень точности равную 2.

Реализуем программно метод, вычисляющий эту сумму при заданных параметрах $f(x), a, b, h$.

```python
def trapezoids(f, a, b, h):
    I = (f(a) + f(b))/2
    N = int((b-a)/h)
    x_k = a
    for k in range(N):
        I += f(a + k * h)
    I *= h
    return I
```

Итак, задавая обеим формулам одни и те же значения точности $\varepsilon$, мы можем сравнивать для них скорость число делений отрезка $[a,b]$ и время работы алгоритма для каждой из формул.
