```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

1. Загрузить данные из файла train.csv. Подготовить данные для дальнейшего анализа, исключив их них экстремально большие значения.

```python
dataset = pd.read_csv('train.csv')
dataset.head()
```

```python
data = dataset['LotArea']
data
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(data)
plt.show()
```

```python
Q3 = data.describe()['75%']
Q1 = data.describe()['25%']
```

```python
upper_bound = Q3 + 1.5 * (Q3 - Q1)
upper_bound
```

```python
data = data[data < upper_bound]
data
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(data)
plt.show()
```

2. Подобрать три непрерывных распределения (минимум два должны
подходить по форме) и с помощью каждого из них приблизить
соответствующие распределения выборочных данных.

```python
plt.figure(figsize=(16, 8))
sns.histplot(data, kde=True)
plt.show()
```

```python
bins = np.linspace(data.min(), data.max(), data.size)

mu, std = stats.norm.fit(data)
norm_distribution = stats.norm.pdf(bins, mu, std)
print('Нормальное распределение ~', mu, std)

a, mu, std = stats.alpha.fit(data)
alpha_distribution = stats.alpha.pdf(bins, a, mu, std)
print('Альфа распределение ~', a, mu, std)

dfn, dfd, mu, std = stats.f.fit(data)
f_distribution = stats.f.pdf(bins, dfn,dfd, mu, std)
print('F распределение ~', dfn, dfd, mu, std)

a, mu, std = stats.gamma.fit(data)
gamma_distribution = stats.gamma.pdf(bins, a, mu, std)
print('Гамма распределение ~', a, mu, std)
```

```python
plt.figure(figsize=(16, 8))
sns.histplot(data, bins=30, kde=True, stat='density', label='Исходное распределение')
plt.plot(bins, norm_distribution, 'm-', label='Нормальное распределение')
plt.plot(bins, alpha_distribution, 'r-', label='Альфа распределение')
plt.plot(bins, f_distribution, 'g-', label='F распределение')
plt.plot(bins, gamma_distribution, 'b-', label='Гамма распределение')
plt.legend()
plt.show()
```

```python
k = 10
observed_frequencies, interval_edges = np.histogram(data, k)

def expected_frequencies(dist, params, interval_edges):
    expected = []
    for i in range(k):
        expected.append(dist.cdf(interval_edges[i+1], *params) - dist.cdf(interval_edges[i], *params))
    return expected


mu, std = stats.norm.fit(data)
expected_norm = expected_frequencies(stats.norm, (mu, std), interval_edges)

a, mu, std = stats.alpha.fit(data)
expected_alpha = expected_frequencies(stats.alpha, (a, mu, std), interval_edges)

dfn, dfd, mu, std = stats.f.fit(data)
expected_f = expected_frequencies(stats.f, (dfn, dfd, mu, std), interval_edges)

a, mu, std = stats.gamma.fit(data)
expected_gamma = expected_frequencies(stats.gamma, (a, mu, std), interval_edges)
```

```python
observed_frequencies
```

```python
def chi_square_stat(observed, expected):
    n = data.size
    statistic = np.sum((observed - n*np.array(expected))**2 / (n*np.array(expected)))
    return statistic
```

```python
df = k - 2 - 1
stat = chi_square_stat(observed_frequencies, expected_norm)
print('Хи-квадрат нормального распределения: stat=' + str(stat) + ', P=' + str(1 - stats.chi2.cdf(stat, df)))

df = k - 3 - 1
stat = chi_square_stat(observed_frequencies, expected_alpha)
print('Хи-квадрат альфа распределения: stat=' + str(stat) + ', P=' + str(1 - stats.chi2.cdf(stat, df)))

df = k - 4 - 1
stat = chi_square_stat(observed_frequencies, expected_f)
print('Хи-квадрат F распределения: stat=' + str(stat) + ', P=' + str(1 - stats.chi2.cdf(stat, df)))

df = k - 3 - 1
stat = chi_square_stat(observed_frequencies, expected_gamma)
print('Хи-квадрат гамма распределения: stat=' + str(stat) + ', P=' + str(1 - stats.chi2.cdf(stat, df)))
```

```python
mu, std = stats.norm.fit(data)
norm = stats.norm.rvs(mu, std, data.size)
expected_frequencies, interval_edges = np.histogram(norm, k)
stat, p, df, expected = stats.chi2_contingency([observed_frequencies, expected_frequencies])
print('Хи-квадрат нормального распределения: stat=' + str(stat) + ', P=' + str(p))

a, mu, std = stats.alpha.fit(data)
alpha = stats.alpha.rvs(a, mu, std, data.size)
expected_frequencies, interval_edges = np.histogram(alpha, k)
stat, p, df, expected = stats.chi2_contingency([observed_frequencies, expected_frequencies])
print('Хи-квадрат альфа распределения: stat=' + str(stat) + ', P=' + str(p))

dfn, dfd, mu, std = stats.f.fit(data)
f = stats.f.rvs(dfn,dfd, mu, std, data.size)
expected_frequencies, interval_edges = np.histogram(f, k)
stat, p, df, expected = stats.chi2_contingency([observed_frequencies, expected_frequencies])
print('Хи-квадрат f распределения: stat=' + str(stat) + ', P=' + str(p))

a, mu, std = stats.gamma.fit(data)
gamma = stats.gamma.rvs(a, mu, std, data.size)
expected_frequencies, interval_edges = np.histogram(gamma, k)
stat, p, df, expected = stats.chi2_contingency([observed_frequencies, expected_frequencies])
print('Хи-квадрат гамма распределения: stat=' + str(stat) + ', P=' + str(p))
```
