```python
def calculate_statistics(numerical_features):
    descriptive_statistics = numerical_features.describe()
    descriptive_statistics.loc['median'] = numerical_features.median()
    descriptive_statistics.loc['mode'] = numerical_features.mode().loc[0]
    descriptive_statistics.loc['IQR'] = descriptive_statistics.loc[['75%']].iloc[0] - descriptive_statistics.loc[['25%']].iloc[0]
    descriptive_statistics.loc['skewness'] = skew(numerical_features, bias=False)
    descriptive_statistics.loc['kurtosis'] = kurtosis(numerical_features, bias=False)
    return descriptive_statistics
```

```python
sns.histplot(x=data['usd_pred'], kde=True)
plt.show()
```

```python
calculate_statistics(data['usd_pred'])
```

3. Выдвинуть 2 гипотезы относительно влияния фактора на бинарную
метрику. С помощью Биномиального теста (логику теста реализовать
самому) проверить гипотезы, дать интерпретацию полученных
результатов.

```python
def binomial_test(k, n, p=0.5):
    p_value = sum(comb(n, i) * (p**i) * ((1-p)**(n-i)) for i in range(k, n+1))
    return p_value * 2 
```

Предположим, что каждый студент с некоторой вероятностью $p$ имеет работу.

Студентов можно разделить на 2 группы по половому признаку: $X_1$ – мужчины, $X_2$ – женщины.

Выдвинем гипотезу $$H_0=\{\text{Фактор «мужчина/женщина» не влияет на «наличие работы».}\}$$

```python
male_students = data[data['sex'] == 'М']
female_students = data[data['sex'] == 'Ж']

binomial_test(len(male_students), len(data), p = working_ratio)
```

Предположим, что каждый студент с некоторой вероятностью $p$ часто посещает пары (частота 7+).

Студентов можно разделить на 2 группы по половому признаку: $X_1$ – мужчины, $X_2$ – женщины.

Выдвинем гипотезу $$H_0=\{\text{Фактор «мужчина/женщина» не влияет на «частоту посещения».}\}$$

```python
binomial_test(len(male_students), len(data), p = visit_ratio)
```

4. Выдвинуть 2 гипотезы относительно влияния фактора на непрерывную
метрику. С помощью t-критерия Стьюдента проверить гипотезы, дать
интерпретацию полученных результатов.

Выдвинем гипотезу $$H_0=\{\text{Фактор «наличие работы» не влияет на «случайное число».}\}$$

```python
working_students = data[data['work_status'].str.contains('Да')]['random_value']
non_working_students = data[data['work_status'].str.contains('Нет')]['random_value']

t_stat_work, p_value_work = ttest_ind(working_students, non_working_students, equal_var=False)
p_value_work
```

Выдвинем гипотезу $$H_0=\{\text{Фактор «пол» не влияет на «случайное число».}\}$$

```python
male_students = data[data['sex'] == 'М']['random_value']
female_students = data[data['sex'] == 'Ж']['random_value']

t_stat_sex, p_value_sex = ttest_ind(male_students, female_students, equal_var=False)
p_value_sex
```
